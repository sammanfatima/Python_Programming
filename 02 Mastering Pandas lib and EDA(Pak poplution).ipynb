### **How rural population increased from 1998 to 2017**

```python
pop_rural_1998 = df['POPULATION 1998 (RURAL)'].sum()
pop_rural_2017 = df['ALL SEXES (RURAL)'].sum()

pop_change = pop_rural_2017 - pop_rural_1998
print(pop_change)
```

```python
print("The total rural population of Pakistan have been increased by 2017 as compared to 1998 is as follows:", pop_change)
```

```python
pop_percentage = (pop_change / pop_rural_1998)*100
print(pop_percentage)
```

```python
pop_urban_1998 = df['POPULATION 1998 (URBAN)'].sum()
pop_urban_2017 = df['ALL SEXES (URBAN)'].sum()

pop_change = pop_urban_2017 - pop_urban_1998
print(pop_change)

print("The total URBAN population of Pakistan have been increased by 2017 as compared to 1998 is as follows:", pop_change)
```
