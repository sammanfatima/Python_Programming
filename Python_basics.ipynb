## ***Control Flow loops***

### ***For loops***

```python
i = 0
for i in range(11):
  print(i)
```

```python
n = int(input("enter a number: "))
for i in range(1, 11):
  print(n, "x", i, "=", n*i )
```

## ***Break with for loop***

```python
for letter in "pakistan":
  if letter == "t":
    break
  print(letter)
```

## ***Continue with for loop***

```python
for letter in "pakistan":
  if letter == "t":
    continue
  print(letter)
```

## ***While loop***

```python
i = 0
while i < 6:
  i += 1
  print(i)
```
